```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree  import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

```python
import pandas as pd
column_names=["sepal length","sepal_width","petal_length","petal width","species"]
url="/content/iris.data"
iris_data=pd.read_csv(url,header=None,names=column_names)
print(iris_data.head())
print("shape of the dataframe:",iris_data.shape)
print("number of rows:",iris_data.shape[0])
print ("number of columns:",iris_data.shape[1])
print("\nMemory usage of the dataset:")
print(iris_data.memory_usage(deep=True))
print("total memory usage in bytes:",iris_data.memory_usage(deep=True).sum())
#printing a brief info about the dataset
print("information about the datset:")
iris_data.info()
```

```python
print(iris_data.describe())
print(iris_data.isnull().sum())
```

```python
sns.pairplot(iris_data,hue="species")
plt.show()
# Dropping the 'species' column before calculating correlation
numerical_data = iris_data.drop("species", axis=1)
correlation_matrix=numerical_data.corr()

sns.heatmap(correlation_matrix,annot=True)
plt.show()

#splitting the datset into training and testing model(preprocessing  )
X=iris_data.drop("species",axis=1)
Y=iris_data["species"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=10,random_state=50)
```

```python

# Example: Sepal Length vs. Sepal Width
sns.jointplot(x='sepal length', y='sepal_width', data=iris_data, hue='species') # Changed 'sepal width' to 'sepal_width'
plt.show()
```

```python

clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(Y_test, Y_pred))
print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
clf=DecisionTreeClassifier()
clf.fit(X_train,Y_train)
Y_pred=clf.predict(X_test)
accuracy=accuracy_score(Y_test,Y_pred)
print("Accuracy:",accuracy)
print("Classification Report:\n",classification_report(Y_test,Y_pred))
print("Confusion Matrix:\n",confusion_matrix(Y_test,Y_pred))
```
